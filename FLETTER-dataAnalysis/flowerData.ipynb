```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL 설정
url = "https://www.kotjumun.com/shop/shopbrand.html?type=X&xcode=014&mcode=002&sort=&page=1"

# HTML 가져오기
response = requests.get(url)
html = response.content

# HTML 파싱
soup = BeautifulSoup(html, 'html.parser')

# 데이터 추출
flowers = soup.find_all('ul', class_='info') # 꽃이름, 가격
images = soup.find_all('img', class_='MS_prod_img_l') # 꽃 이미지 주소

flower_data = []

for flower in flowers:
    name_tag = flower.find('li', class_='name')
    price_tag = flower.find('li', class_='price')
    
    # 각각의 태그가 존재하는지 확인
    name = name_tag.get_text(strip=True) if name_tag else 'No Name'
    price = price_tag.get_text(strip=True) if price_tag else 'No Price'

    flower_data.append({
        'name': name,
        'price': price,
    })

# 이미지 주소 추출 및 병합
for idx, img in enumerate(images):
    img_src = img['src']
    if idx < len(flower_data):
        flower_data[idx]['img_src'] = "https://www.kotjumun.com" + img_src
    else:
        # 추가적인 이미지가 있을 경우
        flower_data.append({
            'name': 'No Name',
            'price': 'No Price',
            'img_src': "https://www.kotjumun.com" + img_src
        })

# DataFrame으로 변환
df = pd.DataFrame(flower_data)

# CSV 파일로 저장
df.to_csv('flowers.csv', index=False)

print("CSV 파일이 성공적으로 저장되었습니다.")
```

```python
def clean_price(price_str):
    price_str = price_str.replace(',', '') # 쉼표 제거
    price_str = price_str.replace('원', '') # '원' 제거
    try:
        return int(price_str) # 정수형으로 변환
    except ValueError:
        return None # 변환 불가능한 경우 처리

# price 컬럼 값 변환
df['price'] = df['price'].apply(clean_price)
```
